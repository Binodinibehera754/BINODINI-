```python

import dash
from dash import dcc, html, Input, Output
import pandas as pd
import plotly.express as px
import requests

# Load COVID-19 data from the API
def fetch_covid_data():
    url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
    df = pd.read_csv(url)
    df = df[df['continent'].notna()]  # Exclude world aggregates
    df['date'] = pd.to_datetime(df['date'])
    return df

df = fetch_covid_data()
countries = df['location'].unique()

# Initialize Dash app
app = dash.Dash(__name__)
app.title = "COVID-19 Dashboard"

# App layout
app.layout = html.Div([
    html.H1("🌍 COVID-19 Data Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Country:"),
        dcc.Dropdown(
            id='country-dropdown',
            options=[{'label': c, 'value': c} for c in countries],
            value='India'
        )
    ], style={'width': '40%', 'margin': 'auto'}),

    html.Br(),

    dcc.Graph(id='cases-line-chart'),
    dcc.Graph(id='deaths-line-chart'),

    html.Div([
        html.H4("Latest Global Summary"),
        dcc.Graph(id='global-summary')
    ])
])

# Callback to update line charts
@app.callback(
    [Output('cases-line-chart', 'figure'),
     Output('deaths-line-chart', 'figure'),
     Output('global-summary', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_charts(country):
    country_df = df[df['location'] == country]

    # Line chart for new cases
    fig_cases = px.line(
        country_df, x='date', y='new_cases',
        title=f'Daily New COVID-19 Cases in {country}',
        labels={'new_cases': 'New Cases'}
    )

    # Line chart for new deaths
    fig_deaths = px.line(
        country_df, x='date', y='new_deaths',
        title=f'Daily New COVID-19 Deaths in {country}',
        labels={'new_deaths': 'New Deaths'}
    )

    # Global bar chart of total cases by country (latest date)
    latest_date = df['date'].max()
    latest_data = df[df['date'] == latest_date]
    top_countries = latest_data.nlargest(15, 'total_cases')
    fig_summary = px.bar(
        top_countries,
        x='location',
        y='total_cases',
        title=f"Top 15 Countries by Total COVID-19 Cases as of {latest_date.date()}",
        labels={'total_cases': 'Total Cases'},
        color='location'
    )

    return fig_cases, fig_deaths, fig_summary
    import dash
from dash import dcc, html, Input, Output
import pandas as pd
import plotly.express as px
import requests

# Load COVID-19 data from the API
def fetch_covid_data():
    url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
    df = pd.read_csv(url)
    df = df[df['continent'].notna()]  # Exclude world aggregates
    df['date'] = pd.to_datetime(df['date'])
    return df

df = fetch_covid_data()
countries = df['location'].unique()

# Initialize Dash app
app = dash.Dash(__name__)
app.title = "COVID-19 Dashboard"

# App layout
app.layout = html.Div([
    html.H1("🌍 COVID-19 Data Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Country:"),
        dcc.Dropdown(
            id='country-dropdown',
            options=[{'label': c, 'value': c} for c in countries],
            value='India'
        )
    ], style={'width': '40%', 'margin': 'auto'}),

    html.Br(),

    dcc.Graph(id='cases-line-chart'),
    dcc.Graph(id='deaths-line-chart'),

    html.Div([
        html.H4("Latest Global Summary"),
        dcc.Graph(id='global-summary')
    ])
])

# Callback to update line charts
@app.callback(
    [Output('cases-line-chart', 'figure'),
     Output('deaths-line-chart', 'figure'),
     Output('global-summary', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_charts(country):
    country_df = df[df['location'] == country]

    # Line chart for new cases
    fig_cases = px.line(
        country_df, x='date', y='new_cases',
        title=f'Daily New COVID-19 Cases in {country}',
        labels={'new_cases': 'New Cases'}
    )

    # Line chart for new deaths
    fig_deaths = px.line(
        country_df, x='date', y='new_deaths',
        title=f'Daily New COVID-19 Deaths in {country}',
        labels={'new_deaths': 'New Deaths'}
    )

    # Global bar chart of total cases by country (latest date)
    latest_date = df['date'].max()
    latest_data = df[df['date'] == latest_date]
    top_countries = latest_data.nlargest(15, 'total_cases')
    fig_summary = px.bar(
        top_countries,
        x='location',
        y='total_cases',
        title=f"Top 15 Countries by Total COVID-19 Cases as of {latest_date.date()}",
        labels={'total_cases': 'Total Cases'},
        color='location'
    )

    return fig_cases, fig_deaths, fig_summary

# Run the Dash app
if __name__ == '__main__':
    # The method has been renamed from run_server to run
    app.run(debug=True)
```
